```python
ted_files = []

for element in os.listdir("datasets"):
    if "ted_data-" in element:
        ted_files.append(element)
ted_files
```

```python
ted_dfs = []

for ted_file in ted_files:
    with open(f"datasets/{ted_file}", "rb") as file:
        ted_data_file = pickle.load(file)
        ted_dfs.append(ted_data_file)
ted_dfs
```

```python
ted_dfs[2]["index"] = ted_dfs[2]["index"]+10
ted_dfs[3]["index"] = ted_dfs[3]["index"]+10
```

```python
concatened = pd.concat(ted_dfs)
concatened
```

```python
concatened.reset_index(inplace=True, drop=True)
```

```python
concatened.to_pickle("datasets/ted_data.pickle")
```
