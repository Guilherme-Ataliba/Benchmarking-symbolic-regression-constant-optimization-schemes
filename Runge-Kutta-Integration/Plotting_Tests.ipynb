```python
import sympy as smp
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from sympy.parsing.sympy_parser import parse_expr
```

```python
expr = parse_expr("0.1745307068609486*cos(3.124746292564876*(1.000122518033892*(-0.0001336198073827764+x)))")
expr.simplify()
```

```python
from sympy.utilities.lambdify import lambdify

x = smp.symbols("x")

func = lambdify([x], expr)
```

```python
data = pd.read_csv("rungekutta_prepared.csv")
x_range = data.x
x_range = np.array(x_range)
```

```python
y = func(x_range)
```

```python
plt.plot(x_range, y, c="blue", linestyle="dashed", label="Predicted", alpha=0.5)
plt.plot(data.x, data.y, c="red", label="Original", alpha=0.5)
# plt.xlim(0, 2)
plt.legend()
plt.show()
```
